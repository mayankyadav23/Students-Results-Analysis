# Data Analysis

```python
plt.figure(figsize=(5,5))
plt.title('Gender Distribution')
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.ylabel('No. of Students')
plt.show()
```

#From the above chart, we have realised that the number of females data is more than males.

```python
gb=df.groupby("ParentEduc").agg({"MathScore" : 'mean',"ReadingScore" : 'mean',"WritingScore" : 'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
plt.title('Scores based on ParentEducation')
sns.heatmap(gb, annot = True)
plt.show()
```

#From the above chart, we have concluded that eduaction of parents have a good impact on their children's scores.

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore" : 'mean',"ReadingScore" : 'mean',"WritingScore" : 'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
plt.title('Scores based on ParentMaritalStatus')
sns.heatmap(gb1, annot = True)
plt.show()
```

#From the above chart, we have concluded that the marital status of parents have negligible impact on their child's scores.

```python
groupA=df.loc[(df['EthnicGroup'] == "group A")].count()
groupB=df.loc[(df['EthnicGroup'] == "group B")].count()
groupC=df.loc[(df['EthnicGroup'] == "group C")].count()
groupD=df.loc[(df['EthnicGroup'] == "group D")].count()
groupE=df.loc[(df['EthnicGroup'] == "group E")].count()

l=["group A", "group B", "group C", "group D", "group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels=l, autopct= "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
# Lunch Type Influence
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='LunchType', y='MathScore')
plt.title('Math Score Distribution by Lunch Type')
plt.show()
```

#If theres a notable difference in scores between students on standard lunch versus free/reduced lunch, it could point to socioeconomic factors affecting academic performance.

```python
# Weekly Study Hours Correlation
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='WklyStudyHours', y='MathScore')

plt.title('Weekly Study Hours vs Math Score')
plt.xlabel('Weekly Study Hours')
plt.ylabel('Math Score')
plt.show()
```

#A strong positive correlation between weekly study hours and scores suggests that more study time leads to better academic performance. This emphasizes the importance of encouraging consistent study habits.

```python
correlation = df[['MathScore', 'ReadingScore', 'WritingScore']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Scores')
plt.show()
```

#A high correlation between scores in Math, Reading, and Writing suggests that students performing well in one subject tend to perform well in others. This reinforces the importance of a well-rounded curriculum.

```python
#Score Distribution
plt.figure(figsize=(15, 5))
for i, subject in enumerate(['MathScore', 'ReadingScore', 'WritingScore'], 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[subject], bins=20, kde=False)  # Disable KDE for a standard histogram
    plt.title(f'Distribution of {subject}')
    plt.xlabel(subject)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
